```python
import cv2
import matplotlib.pyplot as plt

# Option 1: Raw string (add 'r' prefix)
img = cv2.imread(r'D:\IPCV\LAB-EXP-5\images\1.jpeg', 0)

# Option 2: Forward slashes
img = cv2.imread('D:/IPCV/LAB-EXP-5/images/1.jpeg', 0)

# Plot the image
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Image')
plt.show()
```

```python
gauss = img+np.random.normal(0, 25, img.shape)
```

```python
gauss = np.clip(gauss, 0, 255).astype(np.uint8)
```

```python
print(gauss)
```

```python
sp = img.copy()
```

```python
prob = 0.02
rand = np.random.rand(*img.shape)
sp[rand < prob] = 0
sp[rand > 1 - prob] = 255
```

```python
cv2.imwrite('gauss.jpg', gauss)
```
